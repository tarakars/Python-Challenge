```python
import pandas as pd
import tweepy
import json
import requests
from datetime import datetime
import matplotlib.pyplot as plt
from config import consumer_key,consumer_secret,access_token,access_token_secret
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
target_users = ["@BBC", "@CBS", "@CNN", "@FoxNews","@nytimes"]
counter = 1
sentiments = []
```

```python
for user in target_users:
    for x in range(5):
        
        public_tweets = api.user_timeline(user,page = x+1)
        
        for tweets in public_tweets:
            compound = analyzer.polarity_scores(tweets["text"])["compound"]
            pos = analyzer.polarity_scores(tweets["text"])["pos"]
            neu =  analyzer.polarity_scores(tweets["text"])["neu"]
            neg = analyzer.polarity_scores(tweets["text"])["neg"]
            
            
            sentiments.append({"Channel":user,
                           "Date": tweets["created_at"], 
                           "Compound": compound,
                           "Positive": pos,
                           "Negative": neu,
                           "Neutral": neg,
                           "Tweets Ago": counter})
            counter +=1
            
        
```

```python
news_mood_df = pd.DataFrame.from_dict(sentiments)
news_mood_df.sort_values("Date",ascending = True).head()
```

```python
news_mood_BBC_df = news_mood_df.loc[news_mood_df["Channel"]=="@BBC"]
news_mood_BBC_df.sort_values("Date",ascending = True).head()
```

```python
news_mood_CNN_df = news_mood_df.loc[news_mood_df["Channel"]=="@CNN"]
news_mood_CNN_df["Tweets Ago"] = news_mood_CNN_df["Tweets Ago"]-200
news_mood_CNN_df.sort_values("Date",ascending = True).head()
```

```python
news_mood_CBS_df = news_mood_df.loc[news_mood_df["Channel"]=="@CBS"]
news_mood_CBS_df["Tweets Ago"] = news_mood_CBS_df["Tweets Ago"]-100
news_mood_CBS_df.sort_values("Date",ascending = True).head()
```

```python
news_mood_Fox_df = news_mood_df.loc[news_mood_df["Channel"]=="@FoxNews"]
news_mood_Fox_df["Tweets Ago"] = news_mood_Fox_df["Tweets Ago"]-300
news_mood_Fox_df.sort_values("Date",ascending = True).head()
```

```python
news_mood_NY_df = news_mood_df.loc[news_mood_df["Channel"]=="@nytimes"]
news_mood_NY_df["Tweets Ago"] = news_mood_NY_df["Tweets Ago"]-400
news_mood_NY_df.sort_values("Date",ascending = True).head()
```

```python
plt.scatter(news_mood_BBC_df["Tweets Ago"],news_mood_BBC_df["Compound"],color = "blue",label = "BBC")
plt.scatter(news_mood_CBS_df["Tweets Ago"],news_mood_CBS_df["Compound"],color = "green",label = "CBS")
plt.scatter(news_mood_CNN_df["Tweets Ago"],news_mood_CNN_df["Compound"],color = "red",label = "CNN")
plt.scatter(news_mood_Fox_df["Tweets Ago"],news_mood_Fox_df["Compound"],color = "violet", label = "FOX")
plt.scatter(news_mood_NY_df["Tweets Ago"],news_mood_NY_df["Compound"],color = "yellow", label = "FOX")
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

```python
news_mood_sum_df = news_mood_df[["Channel","Compound"]]
news_mood_sum_group_df = news_mood_sum_df.groupby("Channel")
aggregate = news_mood_sum_group_df["Compound"].mean()
Channels = ["B","CB","CN","F","N"]
```

```python
plt.bar(Channels, aggregate, alpha=0.5, align="center",color = ["lightblue", "green", "red", "blue", "yellow"])
plt.show()
```
