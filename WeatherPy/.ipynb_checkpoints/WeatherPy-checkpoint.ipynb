```python
#importing libraries
from citipy import citipy
import matplotlib.pyplot as plt
import pandas as pd
import random
import requests
from pylab import figure
from config import api_key
import openweathermapy.core as owm
```

```python
city = citipy.WORLD_CITIES_DICT #using citipy library getting all the cities in the world as a dictionary
```

```python
unit = "imperial" # declaring this variable to use this in URL for the unit format I need
cities=[] # creating the cities array to store the list of lattitudes generated randomly.
city_names= [] # creating the city_names array to store the list of city names
country_codes =[] # Creating the country_codes array to store the country codes
latitudes = [] #using to store the lattitudes of each city
longitudes = [] # Using this array to store longitudes of each city
temperature = [] # used to sotre the temperatures
humidity = []# used to sotre the humidity
wind_speed = [] # used to sotre the wind_speed
clouds = [] # used to sotre the clouds
coordinates_list = list(city.keys()) # creating a list of city latitudes from the citipy.WORLD_CITIES_DICT

#creating a loop to get random 500 cities

for i in range(500):
    random_city= random.choice(coordinates_list) # generating a random city from the list of coordinates created
    cities.append(random_city) # appending the randomly generated city in to the cities array
    near_by_city= citipy.nearest_city(*cities[i]) # getting a nearby city of the latitude and longitude generated randomly.
    country_codes.append(near_by_city.country_code)#appending the city code of the nearby city we got above
    city_names.append(near_by_city.city_name)#appending the city name of the nearby city we got above
    lat,lon = cities[i] # getting the latitude and longitude from the co-ordinate we stored fro each city.
    latitudes.append(lat) # appending the latitude to the array created above
    longitudes.append(lon)# appending the longitude to the array created above
    url = "http://api.openweathermap.org/data/2.5/weather?" # this is the base URL for the openweathermapy API
    #query_url = url + "lat=" + str(latitudes[i]) + "&lon=" + str(longitudes[i])+"&units="+unit+"&APPID=" + api_key
    query_url = f"{url}lat={latitudes[i]}&lon={longitudes[i]}&units={unit}&appid={api_key}" # creating a URL with API Key,latitude and longitude to get the weather details for the particular city
    weather_response = requests.get(query_url) # getting the response from the URL built above
    weather_json = weather_response.json()# creating the response we got in to JSON format
    print(f"we are looking in to data for: {city_names[i]}.") # printing the city name we are looking in to.
    print(query_url) # printing the URL we re using to get the responses
    temperature.append(weather_json["main"]["temp_max"]) # getting temperature for particular city from the response which was converted to JSON
    humidity.append(weather_json["main"]["humidity"])# getting Humidity for particular city from the response which was converted to JSON
    wind_speed.append(weather_json["wind"]["speed"])# getting Wind Speed for particular city from the response which was converted to JSON
    clouds.append(weather_json["clouds"]["all"])# getting cloud details for particular city from the response which was converted to JSON
```

```python
#creating a data frame with the data we collected from the Openweathermapy and the related city names and country codes
cities_df = pd.DataFrame({"City":city_names,
                 "Country Code":country_codes,
                 "Latitude":latitudes ,
                 "Longitude":longitudes,"Temperature":temperature,
                 "Humidity":humidity,
                 "Cloudiness":clouds,
                 "Wind Speed":wind_speed}, 
                  columns = ["City","Country Code","Latitude","Longitude","Temperature","Humidity","Cloudiness","Wind Speed"])
cities_df.to_csv("Output.csv", sep=',', encoding='utf-8') # storing the data frame in the form of Excel with the name Output.csv
cities_df.head() #printing the head of the data frame
```

```python
# creating a scatter plot between latitude vs temperature
plt.scatter(cities_df["Latitude"], cities_df["Temperature"],marker="o", facecolors="blue", edgecolors="black", alpha=0.75)
plt.title("City Latitude vs City Temperature") #creating title
plt.xlabel("City Latitude") # creating the Xlabel
plt.ylabel("Max City Temperature(F)") #creating Ylabel
plt.grid(True) # putting grid in
plt.savefig("City Latitude vs City Temperature.png") #saving the plot as a png
plt.show()# displaying the plot in jupyter notebook
```
