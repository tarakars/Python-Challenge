```python
#Dependencies
import pandas as pd
```

```python
#Read in csv file from raw_data folder
stop_data = pd.read_csv("raw_data/Officer_Traffic_Stops_Charlotte_enhanced.csv")
```

```python
#Display DataFrame
stop_data.head()
```

```python
#Drop unwanted columns
stop_data_clean = stop_data.drop(["ObjectID","CreationDate","Creator","EditDate","Editor"], axis=1)

#Display clean Dataframe
stop_data_clean.head()
```

```python
#Export DataFrame as clean csv files
stop_data_clean.to_csv("clean_data/CMPD_Traffic_Stop_Data_enhanced.csv", index=False)
```

```python
#Group by CMPD Division & count number of stops for each division
data = stop_data_clean.set_index("CMPD_Division").groupby(["CMPD_Division"])["Month_of_Stop"].count()

#Create Dataframe
stop_count = pd.DataFrame(data)

#Rename column
stop_count.rename(columns= {"Month_of_Stop" : "Number_of_Stops"}, inplace=True)

#Display Dataframe
stop_count
```

```python
#Calculate total number of stops
total_stops = stop_count["Number_of_Stops"].sum()

#Display
total_stops
```
